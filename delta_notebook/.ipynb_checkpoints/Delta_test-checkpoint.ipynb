```scala
import $ivy.`org.scala-lang:scala-library:2.13.12`

// Add Delta Lake dependency
import $ivy.`io.delta:delta-spark_2.13:3.1.0`

// Add Spark SQL dependency
import $ivy.`org.apache.spark:spark-sql_2.13:3.5.1`
import $ivy.`org.apache.spark:spark-core_2.13:3.5.1`

//hadoop hdfs
import $ivy.`org.apache.hadoop:hadoop-common:3.3.6`
import $ivy.`org.apache.hadoop:hadoop-client:3.3.6`
import $ivy.`org.apache.hadoop:hadoop-client-api:3.3.6`
import $ivy.`org.apache.hadoop:hadoop-hdfs:3.3.6`
```

```scala
import io.delta.tables.DeltaTable
import org.apache.spark.sql.{Encoders, SparkSession}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._
```

```scala
val spark = SparkSession.builder()
  .appName("Jupyter Scala with Delta Lake")
  .master("local[*]")
  .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
  .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
  .getOrCreate()
```

```scala
val delPath="/home/avyuthan-shah/Desktop/dataF"
val version=DeltaTable.forPath(spark, delPath).history(1).select("version").as[Long](Encoders.scalaLong).head 
spark.read.format("delta").option("header","true").option("versionAsOf",version).load(delPath).createOrReplaceTempView("delta_table")
```

```scala
val startTime = System.nanoTime()
val result = spark.sql(
    s"""
    SELECT
    AccountNo,
    DATE,
    TRANSACTIONDETAILS,
    CHQNO,
    VALUEDATE,
    WITHDRAWALAMT,
    DEPOSITAMT,
    BALANCEAMT,
    SUM(WITHDRAWALAMT) OVER(PARTITION BY AccountNo ORDER BY VALUEDATE) AS TotalWithdrawals,
    SUM(DEPOSITAMT) OVER(PARTITION BY AccountNo ORDER BY VALUEDATE) AS TotalDeposits,
    LAG(BALANCEAMT) OVER(PARTITION BY AccountNo ORDER BY VALUEDATE) AS PreviousBalance,
    BALANCEAMT - LAG(BALANCEAMT) OVER(PARTITION BY AccountNo ORDER BY VALUEDATE) AS BalanceChange
FROM
    delta_table
WHERE
    AccountNo = "409000611074'"
ORDER BY
    VALUEDATE;

      """.stripMargin)
val endTime = System.nanoTime()
```

```scala
result.show(truncate=false)
```

```scala
val elapsedTime = (endTime - startTime) / 1e9 // Time in seconds
```

```scala
val schema = StructType(Array(
      StructField("AccountNo", StringType, nullable = false),
      StructField("DATE", StringType, nullable = false),
      StructField("TRANSACTIONDETAILS", StringType, nullable = false),
      StructField("CHQNO", StringType, nullable = true),
      StructField("VALUEDATE", StringType, nullable = false),
      StructField("WITHDRAWALAMT", DoubleType, nullable = true),
      StructField("DEPOSITAMT", DoubleType, nullable = true),
      StructField("BALANCEAMT", DoubleType, nullable = true)
    ))
```

```scala
val df=spark.read
//      .option("header","true")
      .option("treatEmptyValuesAsNulls", "true")
      .option("inferSchema", "false")
      .schema(schema)
      .csv("/home/avyuthan-shah/Desktop/F1Intern/Datasets/banktrans10k.csv")

import spark.implicits._  //to use to_date and get map schemas

    //Filtering Csv file to remove ambiguity during read
    val new_df = df
      .withColumn("AccountNo", regexp_replace($"AccountNo", "'", ""))
      .withColumn("DATE", regexp_replace($"DATE", "'", ""))
      .withColumn("CHQNO", regexp_replace($"CHQNO", "'", ""))
      .withColumn("TRANSACTIONDETAILS", regexp_replace($"TRANSACTIONDETAILS", "'", ""))
      .withColumn("VALUEDATE", regexp_replace($"VALUEDATE", "'", ""))
```

```scala
val filtered_df=new_df
      .withColumn("DATE", to_date($"DATE", "yyyy-MM-dd"))
      .withColumn("VALUEDATE", to_date($"VALUEDATE", "yyyy-MM-dd"))

    filtered_df.show()
```

```scala
val deltaT=DeltaTable.forPath(spark,delPath)

val startTime=System.nanoTime()

deltaT.as("dt")
  .merge(
    filtered_df.as("nd"),
    "dt.AccountNo = nd.AccountNo AND dt.VALUEDATE = nd.VALUEDATE AND dt.BALANCEAMT=nd.BALANCEAMT")
  .whenMatched().updateAll() //Note: To use updateAll the schema should match for both dataframe otherwise use updateExpr with Map() to map schema
  .whenNotMatched().insertAll()//Same condition as updateAll
  .execute()

val endTime = System.nanoTime()
val elapsedTime = (endTime - startTime) / 1e9 // Time in seconds
println()
println(s"Elapsed time to load from csv: $elapsedTime")
```

```scala
spark.stop()
```
