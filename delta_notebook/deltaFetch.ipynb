```scala
val version=DeltaTable.forPath(spark, delPath).history(1).select("version").as[Long](Encoders.scalaLong).head //Get Latest Version
writeFile("true")
val startTime = System.nanoTime()

spark.read.format("delta").option("versionAsOf",2).load(delPath).createOrReplaceTempView("delta_table")
val result10=spark.sql(
s"""
   WITH AccountStats AS (
    SELECT 
        AccountNo, 
        DATE, 
        BALANCEAMT,
        SUM(WITHDRAWALAMT) OVER (PARTITION BY AccountNo ORDER BY DATE ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS RunningWithdrawals,
        SUM(DEPOSITAMT) OVER (PARTITION BY AccountNo ORDER BY DATE ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS RunningDeposits
    FROM delta_table
),
PivotData AS (
    SELECT 
        AccountNo,
        SUM(CASE WHEN TRANSACTIONDETAILS = 'Deposit' THEN DEPOSITAMT ELSE 0 END) AS TotalDeposits,
        SUM(CASE WHEN TRANSACTIONDETAILS = 'Withdraw' THEN WITHDRAWALAMT ELSE 0 END) AS TotalWithdrawals
    FROM delta_table
    GROUP BY AccountNo
),
DeltaValues AS (
    SELECT 
        AccountNo, 
        DATE, 
        BALANCEAMT, 
        BALANCEAMT - LAG(BALANCEAMT, 1, 0) OVER (PARTITION BY AccountNo ORDER BY DATE) AS DeltaValue
    FROM delta_table
)
SELECT 
    a.AccountNo, 
    a.DATE, 
    a.BALANCEAMT, 
    a.RunningWithdrawals, 
    a.RunningDeposits,
    p.TotalDeposits, 
    p.TotalWithdrawals, 
    d.DeltaValue
FROM 
    AccountStats a
LEFT JOIN 
    PivotData p ON a.AccountNo = p.AccountNo
LEFT JOIN 
    DeltaValues d ON a.AccountNo = d.AccountNo AND a.DATE = d.DATE
ORDER BY 
    a.AccountNo, a.DATE;

""".stripMargin)

result10.show(truncate=false)
val endTime = System.nanoTime()
writeFile("false")
val elapsedTime = (endTime - startTime) / 1e9 // Time in seconds
println(s"Elapsed time for retrival: '$elapsedTime' ")
```
